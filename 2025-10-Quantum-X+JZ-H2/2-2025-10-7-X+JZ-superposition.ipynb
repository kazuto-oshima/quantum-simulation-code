```python
#[1]  generate E_{0} and construct {1 \over \sqrt{2}}(E_{0} \pm E_{1})
######## 

from qiskit import *
import numpy as np

n_site=1  ##### 
N=n_site 
J=1
step_max=288
t_max=36
dt=t_max/step_max


### simulator and Shots
simulator=Aer.get_backend('qasm_simulator')
n_shots=10000000 #10^7 trial numbers

##### define quantum circuit
qr=QuantumRegister(6,'qr')
cr=ClassicalRegister(6,'cr')
qc=QuantumCircuit(qr, cr)

### measurement
qc_meas=QuantumCircuit(qr,cr)
qc_meas.measure(qr,cr)

### LIst
time_list_qiskit=[]
magx_list_qiskit=[]
mag_list_qiskit=[]

##### 
qc.x(qr[5])  ### initial states 
       
qc.h(qr[5])
qc_total=qc.compose(qc_meas)
qc.h(qr[5])
    

job=execute(qc_total, simulator, shots=n_shots)
result=job.result()
counts=result.get_counts(qc_total)
       
magx=0
for bit_string, count in counts.items():
    obs=0
    for j in range(n_site):
        obs=obs+(-1)**(int(bit_string[j]))
        if j>=n_site-1: 
            magx=magx+obs*count/n_shots/(n_site)
        else:
            magx=magx
magx_list_qiskit.append(magx)
################################################################################
qc_total2=qc.compose(qc_meas)
    

job2=execute(qc_total2, simulator, shots=n_shots)
result2=job2.result()
counts2=result2.get_counts(qc_total2)
        
mag=0
for bit_string, count in counts2.items():
    obs2=0
    for j in range(n_site):
        obs2=obs2+(-1)**(int(bit_string[j]))
        if j>=n_site-1: 
            mag=mag+obs2*count/n_shots/(n_site)
        else:
            mag=mag
    #time_list_qiskit.append(dt*step)
mag_list_qiskit.append(mag)
###################################################
    
print('X=',magx)   ######  X
print('Z=',mag)  #####   Z
print('EE=',magx+J*mag)
print('[0]-results')
```

```python
# [3]  unitary  Raising operator    construct {1 \over \sqrt{2}}(E_{0} \pm E_{1}) from E_{0}
J=1
from scipy.linalg import polar
from qiskit.circuit.library import UnitaryGate

from qiskit.quantum_info import Operator

U01AP=np.array([[0.70729,-0.70690],
             [-0.70690,-0.70735]])
U01,_=polar(U01AP)   ##### unitarization by polar decomposition

print(U01)

unitary_gate=UnitaryGate(U01,label="CustomGate")
controlled_gate=unitary_gate.control()


#########################   
#  qc.h(qr[0])            
# qc.cz(qr[0],qr[5])       
# qc.ch(qr[0],qr[5])       
# qc.cz(qr[0],qr[5])       
######## 

#### important
# qc.s(qr[0])       ###  qc.s(qr[0]) in Fig.3    construct {1 \over \sqrt{2}}(E_{0} \pm iE_{1}) from E_{0}
##################


######################################
#  unitary transformation
qc.h(qr[0])  ### qr[0]  controle bit
qc.append(controlled_gate,[0,5])   #### qr[5] physical qubit
qc.h(qr[0])

############################## X
qc.h(qr[5])
qc_total=qc.compose(qc_meas)
qc.h(qr[5])
    

job=execute(qc_total, simulator, shots=n_shots)
result=job.result()
counts=result.get_counts(qc_total)
 
magX0=0    
magX1=0
zero=0
ichi=0
temp0=0
temp1=0
for bit_string, count in counts.items():
    obs0=0
    obs1=0
    for j in range(1):
        obs0=((-1)**(int(bit_string[0])))*count
        obs1=((-1)**(int(bit_string[0])))*count
        obs0=obs0*(1-int(bit_string[1]))*(1-int(bit_string[2]))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(1-int(bit_string[5]))
        obs1=obs1*(1-int(bit_string[1]))*(1-int(bit_string[2]))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(int(bit_string[5]))
        #if j>=n_site-1: 
        temp0=(1-(int(bit_string[1])))*(1-(int(bit_string[2])))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(1-int(bit_string[5]))*count
        temp1=(1-(int(bit_string[1])))*(1-(int(bit_string[2])))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(int(bit_string[5]))*count
        magX0=magX0+obs0
        magX1=magX1+obs1
        zero=zero+temp0
        ichi=ichi+temp1
        #else:
           # magx=magx
magX0=magX0/zero
magX1=magX1/ichi
######################################## Y   

qc.rx(0.5*np.pi,qr[5])
qc_total=qc.compose(qc_meas)
qc.rx(-0.5*np.pi,qr[5])
    

job=execute(qc_total, simulator, shots=n_shots)
result=job.result()
counts=result.get_counts(qc_total)
 
magY0=0    
magY1=0
zero=0
ichi=0
temp0=0
temp1=0
for bit_string, count in counts.items():
    obs0=0
    obs1=0
    for j in range(1):
        obs0=((-1)**(int(bit_string[0])))*count
        obs1=((-1)**(int(bit_string[0])))*count
        obs0=obs0*(1-int(bit_string[1]))*(1-int(bit_string[2]))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(1-int(bit_string[5]))
        obs1=obs1*(1-int(bit_string[1]))*(1-int(bit_string[2]))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(int(bit_string[5]))
        #if j>=n_site-1: 
        temp0=(1-(int(bit_string[1])))*(1-(int(bit_string[2])))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(1-int(bit_string[5]))*count
        temp1=(1-(int(bit_string[1])))*(1-(int(bit_string[2])))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(int(bit_string[5]))*count
        magY0=magY0+obs0
        magY1=magY1+obs1
        zero=zero+temp0
        ichi=ichi+temp1
        #else:
           # magx=magx
magY0=magY0/zero
magY1=magY1/ichi
######################################### Z
qc_total2=qc.compose(qc_meas)
    

job2=execute(qc_total2, simulator, shots=n_shots)
result2=job2.result()
counts2=result2.get_counts(qc_total2)
#print(counts2)

mag0=0
mag1=0
zero=0
ichi=0
temp0=0
temp1=0
for bit_string, count in counts2.items():
    obs20=0
    obs21=0
    for j in range(1):
        obs20=((-1)**(int(bit_string[0])))*count
        obs21=((-1)**(int(bit_string[0])))*count
        obs20=obs20*(1-int(bit_string[1]))*(1-int(bit_string[2]))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(1-int(bit_string[5]))
        obs21=obs21*(1-int(bit_string[1]))*(1-int(bit_string[2]))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(int(bit_string[5]))
        temp0=(1-(int(bit_string[1])))*(1-int(bit_string[2]))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(1-int(bit_string[5]))*count
        temp1=(1-(int(bit_string[1])))*(1-int(bit_string[2]))*(1-int(bit_string[3]))*(1-int(bit_string[4]))*(int(bit_string[5]))*count
       #if j>=n_site-1:
        mag0=mag0+obs20
        mag1=mag1+obs21
        zero=zero+temp0
        ichi=ichi+temp1   
       # else:
            #mag=mag
mag0=mag0/zero
mag1=mag1/ichi
###################################################
print('zero=', zero)    
print('Y0=',magY0)   ######  Y    
print('X0=',magX0)   ######  X  
print('Z0=',mag0)  #####   Z　　　　
#print('EE=',magx+J*mag)
print('3-0-results')
#################################################
print('ichi=', ichi)    
print('Y1=',magY1)   ######  Y    
print('X1=',magX1)   ######  X    
print('Z1=',mag1)  #####   Z　
#print('EE=',magx+J*mag)
print('3-1-results')  ### matrix element 
###############################################  matrix-elemnts +
print('Y0-Y1', magY0-magY1)
print('X0-X1', magX0-magX1)
print('Z0-Z1', mag0-mag1)
print('Matrix Elements')
```
